```python
import sys
sys.path.append('../')
sys.path.append('../../../../utilities_ths') # path is where this module is

from utilities_ths import MyLabel

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pickle
import time
from   astropy.time          import Time

from astropy.visualization import quantity_support
import astropy.units as u

from IPython.display import display

from astropy.utils.iers import conf
conf.auto_max_age = None

from grb import GammaRayBurst

# This is the PICKLE file where I dump the grb classes

nmaxgrb = 1000 # max number of gbr in population file
grbpopfile = Path("grbpop_"+str(nmaxgrb)+".bin")

# Styles
#plt.rcdefaults()
#plt.rcParams['font.weight'] = 'bold'
#plt.rcParams['axes.labelweight'] = 'bold'
#plt.rcParams['axes.titleweight'] = 'bold'
plt.style.use('seaborn-talk') # Make the labels readable
#plt.style.use('seaborn-poster') # Make the labels readable
# print(plt.style.available)
```

# Create or read the population file

```python
start_pop = time.time()
if grbpopfile.is_file():
    print(" Population read from ",grbpopfile)
    infile  = open(grbpopfile,"rb")
    grbs    = pickle.load(infile)
    infile.close()
    print("  ... read ",len(grbs)," GRBs")
else:
    print(" Creating population in ",grbpopfile)
    grbs = []
    for i in range(1,nmaxgrb+1):
        if (i <= 10) or (np.mod(i,int(nmaxgrb/10)) == 0): print(i," ",end="")
        file  = Path("../../input/lightcurves/LONG_FITS")/Path("Event"+str(i)+".fits")
        burst = GammaRayBurst.read_new(file)
        grbs.append(burst)
    outfile  = open(grbpopfile,"wb")
    pickle.dump(grbs,outfile)
    outfile.close()
    print("\n ... created")
end_pop = time.time()
elapsed = end_pop-start_pop
print("               Duration   = {0:8.2f} (s)".format(elapsed))
```

# Create a dictionnary of the GRB population

```python
# Get class attributes
burst = GammaRayBurst()
members = [attr for attr in dir(burst) if not callable(getattr(burst, attr)) and not attr.startswith("__")]

for mb in members[:3]: print("Attribute = ",mb," : ",getattr(burst, mb))
```

```python
# Create a dictionnary of the GRB lists (for plots etc.) - note : units are lost because of fluxval
grb = {}
for i in range(len(members)):
    # print(" Scanning : ",members[i])
    values = []
    done = False
    unit = 1
    
    for j,evt in enumerate(grbs):
        x = getattr(evt,members[i])
        #print(x)
        if (isinstance(x, u.Quantity)): 
            values.append(x.value)
        else:
            values.append(x)
    grb[members[i]] = np.asarray(values)
    
```

# Get GRB visible in South and in North

```python
def GRBStat(tag):
    s = []
    n = []
    for g in grb[tag]:
        s.append(g["South"])
        n.append(g["North"])
    n = np.asarray(n)
    s = np.asarray(s)
    
    print("\n General statistics - ", tag)
    print(" North       : ",len(grb["t90"][n]))
    print(" North only  : ",len(grb["t90"][n]) - len(grb["t90"][n*s]))

    print(" South       : ",len(grb["t90"][s]))
    print(" South only  : ",len(grb["t90"][s])- len(grb["t90"][n*s]))
    print(" Both        : ",len(grb["t90"][n*s]))
    print(" Notseen     : ",len(grb["t90"])-len(grb["t90"][n+s]))
    print(" all         : ",len(grb["t90"]))
    return n,s

#------------------
north,south = GRBStat("vis_tonight")
n,s = GRBStat("vis_prompt")
n,s = GRBStat("vis")
```

# Timing

```python
dt_north   = []
dtobs_north = []
dt_south   = []
dtobs_south = []
inorth = 0
isouth = 0
icount = 0

for i in range(len(grb["t90"])):
    t_trig  = grb["t_trig"][i].datetime
    if (grb["vis_tonight"][i]["North"]):
        inorth +=1
        tstart = grb["t_true"][i]["North"][0].datetime
        tstop  = grb["t_true"][i]["North"][1].datetime
        dtobs_north.append((tstop - tstart).total_seconds())
        dt_north.append((tstart  - t_trig).total_seconds())
        
    if (grb["vis_tonight"][i]["South"]):
        isouth +=1
        tstart = grb["t_true"][i]["South"][0].datetime
        tstop  = grb["t_true"][i]["South"][1].datetime
        dtobs_south.append((tstop - tstart).total_seconds())
        dt_south.append((tstart  - t_trig).total_seconds())

dt_north    = np.asarray(dt_north)
dt_south    = np.asarray(dt_south)
dtobs_north = np.asarray(dtobs_north)
dtobs_south = np.asarray(dtobs_south)

print(" North : ",inorth,len(dt_north))
print(" South : ",isouth, len(dt_south))
```

## Delay from prompt

```python
nbin = 100
maskn = (dt_north>0)
masks = (dt_south>0)

fig, (axn,axs) = plt.subplots(nrows=1,ncols=2,figsize=(15,6))
axn.hist(dt_north[maskn]/3600,bins=nbin,color='blue',label=MyLabel(dt_north[maskn],label="North"))
#axn.set_yscale('log')
axn.set_ylabel("Counts")
axn.set_xlabel("Delay since trigger (h)")
axn.legend()

axs.hist(dt_south[masks]/3600,bins=nbin,color='red',label=MyLabel(dt_south[masks],label="South"))
#axs.set_yscale('log')
axs.set_ylabel("Counts")
axs.set_xlabel("Delay since trigger (h)")
axs.legend()

plt.show()
```
