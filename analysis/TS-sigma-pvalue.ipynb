```python
from scipy.stats import chi2,norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
ndof = 1
TS = np.linspace(1,100,10)
```

```python
pvalue  = 1 - chi2.cdf(TS, ndof)
pvalue # Probability that this is wrong 
pvalue1  = 1 - chi2.cdf(TS, ndof+1)
sigma = norm.ppf(1-pvalue) 
```

```python
plt.plot(TS,pvalue,label="pvalue")
plt.plot(TS,pvalue1,label="pvalue ndof+1")
plt.plot(TS,(1-97.72/100)*np.ones(len(TS)),'--',label="2s Li&Ma")
plt.ylabel("pvalue")
plt.xlabel("TS")
plt.semilogy()
plt.legend()
plt.show()
```

```python
plt.plot(TS,sigma,label="sigma")
plt.plot(TS,np.sqrt(TS),label="sqrt TS")
plt.plot(TS,2*np.ones(len(TS)),'--',label="2s Li&Ma")
plt.plot(TS,1*np.ones(len(TS)),'--',label="ratio=1")
plt.plot(TS,np.sqrt(TS)/sigma,label="ratio")
plt.xlabel("TS")
plt.ylabel("Sigma")
plt.legend()
plt.show()
```
