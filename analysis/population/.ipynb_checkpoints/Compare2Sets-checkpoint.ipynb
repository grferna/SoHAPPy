```python
nbin=100
sigmin=5
sigmax=100
mask1 = (g_ana1.sigmx<=sigmax) & (g_ana1.sigmx>sigmin)
mask2 = (g_ana2.sigmx<=sigmax) & (g_ana2.sigmx>sigmin)

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3,figsize=(15,5))

n, bins, p = ax1.hist(g_ana1.sigmx[mask1],bins=nbin,alpha=0.5,
                      label=MyLabel(g_ana1.sigmx[mask1],"old"))
ax1.hist(g_ana2.sigmx[mask2],bins=bins,alpha=0.5,
         label=MyLabel(g_ana2.sigmx[mask2],"new"))
ax1.set_yscale("log")
ax1.legend()

ax2.scatter(np.log10(g_ana1.sigmx),np.log10(g_ana2.sigmx),marker="+")
ax2.plot([-1,3],[-1,3],ls=":",color="red")
ax2.axvline(np.log10(3),label="3s",color="tab:orange",ls=":")
ax2.axhline(np.log10(3),label="3s",color="tab:orange",ls=":")
ax2.axvline(np.log10(5),label="5s",color="tab:green",ls=":")
ax2.axhline(np.log10(5),label="5s",color="tab:green",ls=":")
ax2.set_xlabel("old")
ax2.set_ylabel("new")

ax3.scatter(np.log10(g_ana2.sigmx),100*(g_ana2.sigmx-g_ana1.sigmx)/g_ana1.sigmx,marker="+")
ax3.axvline(np.log10(3),label="3s",color="tab:orange",ls=":")
ax3.axvline(np.log10(5),label="5s",color="tab:green",ls=":")
plt.show()
```
